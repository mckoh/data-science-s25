```python
from pandas import read_csv
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster import hierarchy
from matplotlib import pyplot as plt
```

## Daten Laden

Hier haben wir die Date (1) geladen, (2) nur die nummerischen Spalten herausgegriffen, (3) die Spalte mit der fortlaufenden Nummer ("Unnamed: 0") entfernt, und (4) alle Zeilen mit fehlenden (NaN) Werten gelöscht. Diese Folge von Verarbeitungsschritten könnten wir als **Pipeline** bezeichnen.

```python
data = read_csv("../../data/fake_profiles.csv")   # Daten laden
num_data = data.select_dtypes(include="number")   # Nummerische Daten selektieren
num_data.drop("Unnamed: 0", axis=1, inplace=True) # Unnamed:0-Spalte löschen
num_data.dropna(inplace=True)                     # Zeilen mit leeren Werten löschen
```

## Clustering

### Agglomerativ Hierichisches Clustering

```python
hierarchical_cluster_engine = AgglomerativeClustering(n_clusters=2)
hierarchical_cluster_engine.fit_predict(num_data)
```

### K-Means Clustering

```python
k_means_engine = KMeans(n_clusters=2, max_iter=15)
k_means_engine.fit_predict(num_data)
```

### Scipy Clustering

```python
Z = hierarchy.linkage(num_data)  # Anfertigen einer Distanz-Matrix
hierarchy.dendrogram(Z)          # Darstellen der Distanz-Matrix mit Dendrogram
```
